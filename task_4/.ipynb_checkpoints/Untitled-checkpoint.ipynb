```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df.drop('id', axis= 1, inplace= True)
df.shape

df.info()
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.head()
```

```python
label_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
X = df.drop("stroke", axis=1)
y = df["stroke"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_probs = model.predict_proba(X_test_scaled)[:, 1]
y_probs[:10]
```

```python
y_pred_default = (y_probs >= 0.5).astype(int)
y_pred_default[:10]
```

```python
y_test[:10]
```

```python
cm = confusion_matrix(y_test, y_pred_default)
precision = precision_score(y_test, y_pred_default)
recall = recall_score(y_test, y_pred_default)
roc_auc = roc_auc_score(y_test, y_probs)

print("Confusion Matrix (0.5):")
print(cm)
print("Precision:", precision)
print("Recall:", recall)
print("ROC-AUC:", roc_auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--", label="Random Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
threshold = 0.25
y_pred_025 = (y_probs >= threshold).astype(int)

cm_025 = confusion_matrix(y_test, y_pred_025)
precision_025 = precision_score(y_test, y_pred_025)
recall_025 = recall_score(y_test, y_pred_025)

print("\nConfusion Matrix (0.25):")
print(cm_025)
print("Precision (0.25):", precision_025)
print("Recall (0.25):", recall_025)
```
